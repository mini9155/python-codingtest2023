### 재귀호출(Recursive Call)

```python
def printall(n):
    print(f'n ={n}')
    return n

def plus(x, y):
    answer = x + y
    printall(answer)
    return

plus(3,4)
```

```python
count = 10

def openbox():
    global count
    print('종이박스를 엽니다 >.<')
    count -= 1
    if count == 0:
        print('반지를 넣고 반환합니다******')
        return # 데이터가 있으면 데이터를 들고 호출한 곳으로 돌아가라
    
    openbox() # 자기 자신을 다시 호출 ,부른 만큼 출력
    print('종이 상자를 닫습니다.')
```

```python
openbox() # 함수를 실행하기 위한 호출
```

```python
sumvalue = 0

for n in range(10,0,-1):
    sumvalue += n

print("10+9+8+...+1=", sumvalue)
```

```python
def addnumber(num):
    if num <= 1:
        return 1
    else:
        return num + addnumber(num - 1)

print(addnumber(3))
```

```python
factvalue = 1
for n in range(10,0,-1):
    factvalue *= n

print('10*9*8...*=',factvalue)
```

### 팩토리얼 재귀호출

```python
def factorial(num):
    if num <= 1:
        return 1
    else:
        return num * factorial( num - 1)
    
print('10! = ', factorial(1558))
```

```python
### 우주선 발사
import time
def countdown(n):
    if n == 0:
        print('우주선 발사!!')
    else:
        print(n)
        time.sleep(1)
        countdown(n-1)

countdown(10)
        
```

```python
def printstar(n):
    if n > 0:
        print('★'* n)
        printstar(n-1)

printstar(5)
```

### 구구단 재귀 호출

```python
def gugudan(dan, num):
    print(f'{dan} x {num} = {dan * num}')
    if num < 9:
        gugudan(dan, num+1)

for dan in range(2,10):
    print(f'{dan}단 시작 -------')
    gugudan(dan, 1)
```

```python
tab = ''
def pow(x,n):
    global tab
    tab += ' '
    if n == 0:
        return 1
    print(tab + '%d%d^(%d-%d)'%(x,x,n,1))
    return x * pow(x,n-1)

print('2^4')
print(pow(2,4))
```

```python
def fibonacci(n):
    if n ==0: return 0
    elif n==1: return 1
    else:
        return fibonacci(n - 1) + fibonacci(n-2)
    
print('피보나치 수 -> 0 1', end=' ')
for i in range(2,10):
    print(fibonacci(i), end=' ')
```

```python
def palindrom(pstr): # kayak
    if len(pstr) <= 1: #글자가 한 단어 밖에 없다
        return True
    if pstr[0] != pstr[-1]: #  k == k
        return False
    
    return palindrom(pstr[1:len(pstr)-1])

a = 'kayak'
a[1:len(a)-1] # 1 ~ 3
```
