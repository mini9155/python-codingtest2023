# 최소값 위치 확인 함수

```python
def findminidx(ary):
    minidx = 0 # 배열 0부터
    for i in range(1, len(ary)):
        if (ary[minidx]) > ary[i]: #지정된 값이 배열에 현재값보다 크면
            minidx = i # i로 바꿔준다

    return minidx
```

```python
tests = [55,88,33,77]
print(findminidx(tests))
```

### 개선된 선택 정렬

배열을 두 개 만들 필요 X 값 교환 swap 가능

```python
def selectionsort(ary):
    n = len(ary)
    for i in range(0,n-1): #마지막 인덱스가 n-2
        minidx = i 
        for k in range(i+1,n):# 마지막 인덱스가 n-1
            if(ary[minidx]) > ary[k]:
               minidx = k
        tmp = ary[i] # i는 첫번쨰 값
        ary[i] = ary[minidx] # 최소값으로 변경
        ary[minidx] = tmp

    return ary

# 변수

dataary = [4 ,7 ,9, 3, 2]#[188,162,168,120,20,150,177,105]
print('정렬 전 --> ',dataary)
dataary = selectionsort(dataary)
print('정렬 후 --> ', dataary)
```

```python
def findinsertidx(ary, data):
    findidx = -1
    for i in range(0, len(ary)):
        if ary[i] > data:
            findidx = i
            break
    if findidx == -1:
        return len(ary)
    else:
        return findidx
    
```

```python
testary = []
print('50이 들어갈 위치', findinsertidx(testary, 50))
testary = [33,77,88]
print('55가 들어갈 위치', findinsertidx(testary, 55))
```

### 삽입정렬 개선

```python
def insertionsort(ary):
    n = len(ary)
    for end in range(1, n): # 앞의 값 하나 뺴고 끝까지 반복
        for cur in range(end, 0, -1):
            if ary[cur -1] > ary[cur]:
                tmp = ary[cur -1]
                ary[cur -1] = ary[cur]
                ary[cur] = tmp
    return ary

dataary = [188,162,168,120,20,150,177,105]
print('정렬전 -->',dataary)
dataary = insertionsort(dataary)  
print('정렬전 -->',dataary)
```

### 버블 정렬(Bubble sort)

```python
def bubblesort(ary):
    n = len(ary)
    for end in range(n-1, 0, -1): # n-1번째 부터 0까지 역순
        print('사이클# -->', ary)
        for cur in range(0, end): # 0분터 n-2까지 반복
            if ary[cur] > ary[cur+1]:
                ary[cur],ary[cur+1] = ary[cur+1], ary[cur]

dataary = [188,162,168,120,20,150,177,105]
print('정렬전 -->',dataary)
dataary = bubblesort(dataary)  
print('정렬전 -->',dataary)
```

### 퀵 정렬(Quick sort)

```python
def quicksort(ary):
    n =len(ary)
    if n <= 1: # 정렬갯수가 1개 이하면
        return ary
    
    pivot = ary[n // 2] #중간값을 기준값으로 
    leftary, rightary = [],[]

    for num in ary:
        if num < pivot:
            leftary.append(num)
        elif num > pivot:
            rightary.append(num)

    return quicksort(leftary) + [pivot] + quicksort(rightary) #재귀 호출


dataary = [188,162,168,120,20,150,177,105]
print('정렬전 -->',dataary)  
dataary = quicksort(dataary)
print('정렬전 -->',dataary)
```
